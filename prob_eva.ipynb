```python
import pandas as pd
pd.options.display.max_rows = 100
pd.set_option("precision", 6)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('saved_csv/compare_test_df.csv', index_col=0)
df
```

```python
# Add a column to indicate whether the model predict correctly or not
df.loc[df['Predicted'] != df['Actual'], "Right?"] = "wrong"
df.loc[df['Predicted'] == df['Actual'], "Right?"] = "correct"
```

```python
df.head()
```

```python
df['Right?'].value_counts()
```

```python
# Sub-df for predict wrong
df_wrong = df.loc[df['Right?'] == "wrong"]
df_wrong.head()
```

```python
# Sub-df for predict correctly
df_right = df.loc[df['Right?'] == "correct"]
df_right.head()
```

```python
# sub-df: predict(1)  and correctly
df_right_1 = df_right.loc[df_right['Actual'] == 1]
df_right_1.head()
```

```python
df_right_1.describe()
```

- min is 50.73%

```python
df_right_1.loc[df_right_1['Probabilities']<=0.6].sort_values(by='Probabilities')
```

```python
# sub-df: false negative
df_wrong_1 = df_wrong.loc[df_wrong['Actual'] == 1]
df_wrong_1
```

```python
df_wrong_1.sort_values(by='Probabilities')
```

```python
df_wrong_1.describe()
```

- false negative in the range of 1.192% to 43.45%
